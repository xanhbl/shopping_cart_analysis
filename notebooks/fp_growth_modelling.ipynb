```python
import os
import time
import pandas as pd

from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
# đảm bảo đang ở project root
if os.path.basename(os.getcwd()) == "notebooks":
    os.chdir("..")

BASKET_BOOL_PATH = "data/processed/basket_bool.parquet"

basket_bool = pd.read_parquet(BASKET_BOOL_PATH)

print("Basket shape:", basket_bool.shape)
basket_bool.head()
```

```python
MIN_SUPPORT = 0.005

start_time = time.time()

frequent_itemsets_fp = fpgrowth(
    basket_bool,
    min_support=MIN_SUPPORT,
    use_colnames=True
)

elapsed_time = time.time() - start_time

print("=== FP-Growth Results ===")
print("Min support:", MIN_SUPPORT)
print("Number of frequent itemsets:", len(frequent_itemsets_fp))
print(f"Execution time: {elapsed_time:.2f} seconds")

frequent_itemsets_fp.head()
```

```python
MIN_SUPPORT = 0.005

start_time = time.time()

frequent_itemsets_fp = fpgrowth(
    basket_bool,
    min_support=MIN_SUPPORT,
    use_colnames=True
)

elapsed_time = time.time() - start_time

print("=== FP-Growth Results ===")
print("Min support:", MIN_SUPPORT)
print("Number of frequent itemsets:", len(frequent_itemsets_fp))
print(f"Execution time: {elapsed_time:.2f} seconds")

frequent_itemsets_fp.head()
```

```python
rules_fp = association_rules(
    frequent_itemsets_fp,
    metric="lift",
    min_threshold=1.2
)

rules_fp = rules_fp.sort_values("lift", ascending=False)

print("Number of rules:", len(rules_fp))
rules_fp[["antecedents", "consequents", "support", "confidence", "lift"]].head()
```

Điểm khác biệt chính là việc áp dụng thuật toán FP-Growth để khai thác các tập mục phổ biến. FP-Growth không cần sinh candidate itemsets nên hoạt động hiệu quả hơn trên tập dữ liệu lớn. Sau đó, các luật kết hợp được sinh ra dựa trên chỉ số lift để phục vụ phân tích hành vi mua sắm.

Khi giảm ngưỡng min_support xuống 0.005, số lượng frequent itemsets tăng lên đáng kể. Điều này cho thấy FP-Growth phù hợp hơn với dữ liệu lớn và ngưỡng support nhỏ.
